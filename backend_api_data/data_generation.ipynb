```python

# Define the API endpoint
url = "https://app.planetfwd.com/api/lca/generate"

# Define the headers
init_headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
    "Authorization": plt_key
}

for i in data:
    name = i['item']
# Define the request body
    request_body = {
        "name": name,
        "mass": {
            "amount": i['mass'],
            "unit": "g"
        },
        "description": "A delicious pumpkin pie for the holidays.",
        # "emissionSource": {
        #     "amount": 14,
        #     "unit": "oz"
        # }
    }

# Send the POST request
    response = requests.post(url, json=request_body, headers=init_headers)
    print("Response text:", response)
    id = response.json()['id']
    print("the id is",id)
    url = f"https://app.planetfwd.com/api/lca/{id}/generation_status"
    headers = {
    "accept": "application/json",
    "Authorization": plt_key
    }
    while True:
    
        response = requests.get(url, headers=headers)
        print(response.json())
        if response.json()['complete']:
            break
    print(response.status_code)
    print(response.json()['emissionFactor'])
    new_field = response.json()['emissionFactor']
    i['emissions'] = new_field
    time.sleep(2)
```

```python
#create connection pool and establish connection
connection_pool = pool.SimpleConnectionPool(
    1,  # Minimum number of connections in the pool
    10,  # Maximum number of connections in the pool
    connection_string
)
if connection_pool:
    print("Connection pool created successfully")
else:
    print("connection pool missing")
conn = connection_pool.getconn()
cur = conn.cursor()
```

```python
create_table_query = """ 
CREATE TABLE c_company(
    id VARCHAR(255) PRIMARY KEY,
    item VARCHAR(255) NOT NULL,
    mass INT NOT NULL,
    recyclable BOOLEAN NOT NULL,
    biodegradable BOOLEAN NOT NULL,
    amount FLOAT NOT NULL,
    comments VARCHAR(255) NOT NULL,
    emissions FLOAT NOT NULL
);
"""
# Execute the SQL query
cur.execute(create_table_query)
conn.commit()
```

```python
for entry in data:
    # Generate a unique UUID for each entry
    entry_id = str(uuid.uuid4())
    
    cur.execute('''
        INSERT INTO c_company (id, item, mass, recyclable, biodegradable, amount, comments, emissions) 
        VALUES (%s, %s, %s, %s, %s, %s, %s,%s)
    ''', (
        entry_id, entry['item'], entry['mass'], entry['recyclable'], entry['biodegradable'], entry['amount'], entry['comments'], entry['emissions']
    ))
conn.commit()
```
